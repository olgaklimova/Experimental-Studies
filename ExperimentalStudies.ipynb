##Применение Байесовской оптимизации для подбора гиперпараметров и обучение после

```python
!pip install scikit-optimize
```

```python
from skopt import BayesSearchCV
from lightgbm import LGBMRegressor
from sklearn.model_selection import train_test_split
```

```python
# Загрузка данных
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Определение модели LightGBM
model = LGBMRegressor()

# Определение пространства поиска гиперпараметров
param_space = {
    'learning_rate': (0.01, 1.0, 'log-uniform'),
    'n_estimators': (50, 1000),
    'max_depth': (1, 20),
    'num_leaves': (2, 50),
    'min_child_samples': (1, 20),
    'subsample': (0.1, 1.0, 'uniform'),
    'colsample_bytree': (0.1, 1.0, 'uniform')
}

# Создание объекта BayesSearchCV для байесовской оптимизации
opt = BayesSearchCV(model, param_space, n_iter=50, cv=3, scoring='neg_mean_squared_error', random_state=42)

# Поиск оптимальных гиперпараметров
opt.fit(X_train, Y_train)

# Вывод наилучших гиперпараметров
print("Best hyperparameters:", opt.best_params_)

# Оценка качества модели на тестовой выборке
mse = opt.score(X_test, Y_test)
print("Mean squared error on test data:", -mse)
```

```python
#Сортировка по UID
data = data.sort_values(by='UID')
display(data)
```

```python
#Создание столбца Mean_Elapsed_Seconds со средним значением Elapsed_Seconds по UID
data['Mean_Elapsed_Seconds'] = data.groupby('UID')['Elapsed_Seconds'].transform('mean')
display(data)
```

```python
# Создание столбца Mean_2_Elapsed_Seconds средних значений Mean_Elapsed_Seconds с окном 2
# Создаем DataFrame с значениями столбца Elapsed_Seconds
elapsed_seconds_df = data[['UID', 'Elapsed_Seconds']]

# Рассчитываем скользящее среднее
rolling_mean = data.groupby('UID')['Elapsed_Seconds'].rolling(window=2).mean()
# Преобразуем rolling_mean в DataFrame и сбрасываем мультииндекс
rolling_mean_df = rolling_mean.reset_index(level=0, drop=True)

# Объединяем значения Elapsed_Seconds и скользящие средние в новый DataFrame
result_df = pd.concat([elapsed_seconds_df, rolling_mean_df], axis=1)
result_df.columns = ['UID', 'Elapsed_Seconds', 'Rolling_Mean']

# Фильтруем DataFrame для выбранного UID (например, UID=458)
uid_filtered_df = result_df[result_df['UID'] == 458]

# Устанавливаем максимальное количество строк для вывода без сокращения
#pd.set_option('display.max_rows', None)

# Выводим DataFrame
print(uid_filtered_df)
```

```python
pd.set_option('display.max_rows', 10)
# Добавляем скользящее среднее в исходный DataFrame
data['Mean_2_Elapsed_Seconds'] = rolling_mean_df
# Выводим обновленный DataFrame
display(data)
```

```python
# Создание столбцов с различными размерами окон средних
# Создаем DataFrame со значениями столбца Elapsed_Seconds
elapsed_seconds_df = data[['UID', 'Elapsed_Seconds']]

# Размеры окон для скользящих средних
window_sizes = [2, 4, 8, 16, 32, 64]

for window_size in window_sizes:
    # Рассчитываем скользящее среднее
    rolling_mean = data.groupby('UID')['Elapsed_Seconds'].rolling(window=window_size, min_periods=1).mean()
    # Преобразуем rolling_mean в DataFrame
    rolling_mean_df = rolling_mean.reset_index(level=0, drop=True)
    # Добавляем столбец со скользящим средним в исходный DataFrame
    data[f'Mean_{window_size}_Elapsed_Seconds'] = rolling_mean_df

# Выводим обновленный DataFrame
display(data)
```

```python
# Разделение на тестовую и обучающую выборки
# Вычислить количество строк для test_data, чтобы оно составляло 20% от всего набора данных df
test_size = int(len(data) * 0.15)
test_data = pd.DataFrame()
train_data = data

while len(test_data) < test_size:
    # Выберите случайное значение из столбца 'UID'
    random_uid = train_data['UID'].sample().iloc[0]
    num_rows_with_uid = len(data[data['UID'] == random_uid])

    # Если количество строк не превышает test_size, добавляем в тест
    if num_rows_with_uid <= 0.3*test_size:
        test_data = test_data.append(train_data[train_data['UID'] == random_uid])
        # Удалите выбранные строки из исходного датафрейма
        train_data = train_data.drop(train_data[train_data['UID'] == random_uid].index)
    else:
        random_uid = train_data['UID'].sample().iloc[0]
        num_rows_with_uid = len(train_data[train_data['UID'] == random_uid])
        continue


Y_train = train_data.iloc[:, 20]
Y_test = test_data.iloc[:, 20]
X_train = train_data.drop('Elapsed_Seconds', axis=1)
X_test = test_data.drop('Elapsed_Seconds', axis=1)
```

```python
#Обучение модели LightGBM
train_data = lgb.Dataset(X_train, label=Y_train)
test_data = lgb.Dataset(X_test, label=Y_test)

param = {'objective': 'regression', 'metric': 'mse'}
num_round = 5000
lgb_model = lgb.train(param, train_data, num_round, valid_sets=[test_data])
```

```python
#Предсказание и оценка модели на тестовой выборке
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

Y_pred = lgb_model.predict(X_test)

mse = mean_squared_error(Y_test, Y_pred)
print("Mean Squared Error:", mse)

r2 = r2_score(Y_test, Y_pred)
print("R2 score:", r2)
```
